```python
def inc(a):
    return a + 1
def add(a, b):
    return a + b
def mul(a, b):
    return a*b
```

```python
from dask import delayed
```

```python
a = 5; b = 5
```

```python
x = (delayed)(inc)(a=a)
y = (delayed)(inc)(b)
z = (delayed)(add)(x, y)
m = (delayed)(mul)(y, z)
```

```python
m.compute()
```

```python
m.visualize()
```
